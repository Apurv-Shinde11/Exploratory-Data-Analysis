```python
import pandas as pd
from scipy.io import arff
from google.colab import files
uploaded = files.upload()
```

```python
data = arff.loadarff('phpZrCzJR.arff')
df = pd.DataFrame(data[0])
```

```python
df.head()
df.info()
df.describe()
print(df.columns)
```

```python
df['Target'] = df['Target'].str.decode('utf-8')
df.isnull().sum()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

sns.countplot(data=df, x='Target')
plt.title("Class Distribution")
plt.xticks(rotation=45)
plt.show()
```

```python
sns.boxplot(x='Target', y='V2', data=df)
plt.title('Boxplot by Category')
plt.show()
```

```python
df['Target'] = df['Target'].map({'Anomaly': 1, 'Normal': 0})
corr = df.corr()
plt.figure(figsize=(50, 40))
sns.heatmap(corr, annot=True, cmap='coolwarm', fmt='.2f', linewidths=8)
plt.title('Correlation Heatmap')
plt.show()
```
